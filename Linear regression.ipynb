```python
import numpy
import matplotlib
from matplotlib import pylab, mlab, pyplot
from random import randint
from sklearn import datasets, linear_model, metrics
```

# Выборка

```python
# Вариант 11

num = 11
rand_noise = randint(1, 20)
print(rand_noise)

data, target, coef = datasets.make_regression(n_samples = num * 10, 
                                              n_features = num, 
                                              n_informative = int(ceil(num / 5)), 
                                              noise = rand_noise, 
                                              coef = True, 
                                              random_state = 2)
```

# Графики

```python
pyplot.figure(figsize(20, 20))

for pl in range(1, num + 1):
    pyplot.subplot(4, 4, pl)
    pyplot.scatter(list(map(lambda x : x[pl - 1], data)), target, color = 'b')
    pyplot.title(pl)
    pyplot.xlabel(f'feature {pl}')
    pyplot.ylabel('target')
```

# Обучение

```python
linear_regression = linear_model.LinearRegression()
linear_regression.fit(data, target)
predictions = linear_regression.predict(data)
```

# Уравнения

```python
print ("Learned regression model")
fstr = ' + '.join([f'{round(linear_regressor.coef_[i], 4)}*x{i + 1}' for i in range(len(linear_regressor.coef_))])
f'y = {round(linear_regression.intercept_, 4)} + {fstr}'
```

```python
print ("True regression model")
fstring = ' + '.join([f'{round(coef[i], 4)}*x{i + 1}' for i in range(len(coef))])
f'y = {fstring}'
```

# RMSE

```python
mae = metrics.mean_absolute_error(target, predictions)
print ('MAE = ', mae)
mse = metrics.mean_squared_error(target, predictions)
print ('MSE = ', mse)
```
